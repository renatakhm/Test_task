Найдем самый популярный объект в выборке.  
Для этого найдем строку с максимальным значением в стобце 'Total Interactions', который представляет собой сумму действий пользователей над постом - comment, like и share.

```python
dt['Total Interactions'].argmax()
```

```python
dt.iloc[244]
```

Однако стоит учитывать тот факт, что значение столбца 'Total Interactions' зависит от количества людей, которые просмотрели пост. Зачастую чем больше лайков, комментариев и репостов, тем больше была аудитория у данного поста. Поэтому стоит высчитать долю целевых действий('Total Interactions') от общего количества людей, которым виден этот пост ('Lifetime Post Total Reach')

```python
a = dt['Total Interactions']/dt['Lifetime Post Total Reach']
```

```python
a.argmax()
```
